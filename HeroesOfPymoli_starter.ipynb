```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# File to Load (Remember to Change These)
data_file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(data_file)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#The total number of players
len(purchase_data['SN'].unique().tolist())
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#The basic calculations to obtain number of unique items
len(purchase_data['Item Name'].unique().tolist())

# The mean price
average_price = purchase_data["Price"].mean()

# The Max price paid 
max_price = purchase_data["Price"].max()

# The Min price paid
min_price = purchase_data["Price"].min()

#Display the summary data frame
purchase_data_table = pd.DataFrame({"Average Purchase Price":[average_price],"Max Purchase Price":[max_price],"Min Purchase Price":[min_price]})

#The displayed data cleaner formatting
purchase_data_table["Average Purchase Price"] = purchase_data_table["Average Purchase Price"].map("{:.2f}".format)
purchase_data_table.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
player_data = purchase_data.loc[:,["Gender","SN","Age"]]
player_data = player_data.drop_duplicates()
total_players = len(purchase_data['SN'].unique().tolist())
Player_count = player_data["Gender"].value_counts()

player_Gender_per =  player_data["Gender"].value_counts()/total_players * 100

player_Gender_per_table = pd.DataFrame({"Player Count":Player_count,"Player Gender Percentage":player_Gender_per})
player_Gender_per_table.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Purchase_count = purchase_data.groupby(['Gender'])

grouped = purchase_data.groupby('Gender')
name1=grouped['Price'].sum()
name2=grouped['Price'].mean()
name3=grouped['Purchase ID'].count()
name4=grouped['Price'].sum()/Player_count


Purchase_Analysis_Table = pd.DataFrame({"Player Count":name3,"Player average price":name2,"Player total purchase price":name1,"purchase total per person":name4})
Purchase_Analysis_Table = Purchase_Analysis_Table.round(2)
Purchase_Analysis_Table.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9,14,19,24,29,34,39,99]
labels_name=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

player_data["range"] = pd.cut(player_data["Age"],bins , labels=labels_name)

Age_Bin_Total = player_data["range"].value_counts()
Age_Bin_per =  Age_Bin_Total/total_players *100

Age_Bin_df = pd.DataFrame({"Player Count":Age_Bin_Total,"Percentage of Players":Age_Bin_per})
Age_Bin_df = Age_Bin_df.round(2)

Age_Bin_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["range"] = pd.cut(purchase_data["Age"],bins , labels=labels_name)

Pur_Bin_Count = purchase_data.groupby(["range"]).count()["Price"]
Pur_Bin_Total = purchase_data.groupby(["range"]).sum()["Price"]
Pur_Bin_Avg = purchase_data.groupby(["range"]).mean()["Price"]
Pur_Bin_Avg_total_Per = Pur_Bin_Total/Age_Bin_df["Player Count"]


Purchase_Bin_df = pd.DataFrame({"Purchase Count":Pur_Bin_Count,"Average Purchase price":Pur_Bin_Avg,"Total Purchase price":Pur_Bin_Total,"Average Total Purchase per person":Pur_Bin_Avg_total_Per})
Purchase_Bin_df = Purchase_Bin_df.round(2)
Purchase_Bin_df.sort_index()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_Count = purchase_data.groupby(["SN"]).count()["Price"]
SN_Total = purchase_data.groupby(["SN"]).sum()["Price"]
SN_Avg = purchase_data.groupby(["SN"]).mean()["Price"]

SN_df = pd.DataFrame({"Purchase Count":SN_Count,"Average Purchase price":SN_Avg,"Total Purchase price":SN_Total})
SN_df = SN_df.round(2)
SN_df.sort_values("Total Purchase price", ascending =False).head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Item_Count = purchase_data.groupby(["Item ID","Item Name"]).count()["Price"]
Item_Total = purchase_data.groupby(["Item ID","Item Name"]).sum()["Price"]
Item_Price = purchase_data.groupby(["Item ID","Item Name"]).mean()["Price"]


SN_df = pd.DataFrame({"Purchase Count":Item_Count,"Item price":Item_Price,"Total Purchase value":Item_Total})
SN_df = SN_df.round(2)
SN_df.sort_values("Purchase Count", ascending =False).head(5)
```
